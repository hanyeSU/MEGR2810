## Overview of Conditionals and Recursion

### Overview of Conditionals and Recursion
Conditionals allow branching logic using `if`, `elif`, and `else`. Recursion is when a function calls itself.


```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

factorial(5)
```

## Boolean

### Boolean Values
Booleans represent truth values: `True` and `False`.


```python
a = True
b = False
print(a and b)
print(a or b)
print(not a)
```

## Relational Operators

### Relational Operators
Used to compare values: `==`, `!=`, `<`, `>`, `<=`, `>=`


```python
x = 10
y = 20
print(x == y)
print(x < y)
print(x >= y)
```

## Logical Operators

### Logical Operators
`and`, `or`, `not` combine boolean expressions.


```python
x = 5
print(x > 0 and x < 10)
print(x < 0 or x > 3)
print(not(x == 5))
```

## if Statement

### if Statement
Executes a block of code if the condition is true.


```python
x = 7
if x > 5:
    print("x is greater than 5")
```

## else Statement

### else Statement
Provides an alternative block if the condition is false.


```python
x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

## elif Statement

### elif Statement
Checks multiple conditions in sequence.


```python
x = 5
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")
```

## Nested Condition

### Nested Condition
Conditions inside other conditions.


```python
x = 10
if x > 5:
    if x < 15:
        print("x is between 5 and 15")
```

## Recursion

### Recursion Example: Countdown
A function calling itself to count down.


```python
def countdown(n):
    if n <= 0:
        print("Blastoff!")
    else:
        print(n)
        countdown(n - 1)

countdown(5)
```

## while Loop

### while Loop
Repeats a block while a condition is true.


```python
n = 5
while n > 0:
    print(n)
    n -= 1
```
